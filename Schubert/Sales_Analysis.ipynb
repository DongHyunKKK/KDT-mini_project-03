```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
```

```python
file = '../DATA/지역별 IT산업 매출액 통계.xlsx'
```

```python
it_industry_by_region = pd.read_excel(file, header=[0,1], index_col=[0,1,2])
it_industry_by_region.columns.names = [None, None]
it_industry_by_region
```

```python
def year_sum(df, year):
    df_copy = df.copy()
    df_copy = df_copy.xs(year, level=0, axis=1)
    df_copy = df_copy.xs('소계', level=1, axis=0).T
    df_copy.columns = [df_copy.columns[0][0], df_copy.columns[1][0]]
    df_copy['합계'] = df_copy.sum(axis=1)
    return df_copy
```

```python
it_industry_2019 = year_sum(it_industry_by_region, '2019')
it_industry_2020 = year_sum(it_industry_by_region, '2020')
it_industry_2021 = year_sum(it_industry_by_region, '2021')
it_industry_list = [it_industry_2019, it_industry_2020, it_industry_2021]
```

```python
it_industry_list[0].index
```

```python
plt.figure(figsize=(16,6))
for idx, data in enumerate(it_industry_list):
    plt.subplot(1, 3, idx+1)
    plt.pie(data.iloc[1:, 2], labels=data.index[1:], autopct='%1.1f%%', startangle=90)
    plt.title(f'{2019+idx}년 지역별 IT 산업 매출액 비율')
plt.suptitle('2019~2021년 지역별 IT 산업 매출액 비율', fontsize=20)
plt.tight_layout()
plt.show()
```

```python
width = 0.25
multiplier = 0
x = np.arange(len(it_industry_list[0].index))

fig, ax = plt.subplots(figsize=(14,6))

for idx, data in enumerate(it_industry_list):
    offset = width * multiplier
    rects = ax.bar(x + offset, round(it_industry_list[idx]['합계']/10000, 1), width=width, label=f'{2019+idx}년')
    ax.bar_label(rects, padding=3)
    multiplier += 1
ax.set_xlabel('지역', fontsize=15)
ax.set_ylabel('매출액(조)', fontsize=15)
ax.set_xticks(x + width, it_industry_list[0].index, rotation=350)
ax.set_yticks(range(0, 501, 50))
ax.legend(loc='upper right', ncol=3)
plt.suptitle('2019~2021년 지역별 IT 산업 매출액', fontsize=20)
plt.show()
```
